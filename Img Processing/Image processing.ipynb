```python
import numpy as np
import matplotlib.image as img 
import matplotlib.pyplot as plt
from PIL import Image
from statistics import mean 
```

```python
 """ GRAY SCALE """

m = img.imread("graph.png") 

w, h = m.shape[:2] 
  
newImage = np.zeros([w, h, 4]) 
print(w) 
print(h) 
  
for i in range(w): 
   for j in range(h): 
      lst = [float(m[i][j][0]), float(m[i][j][1]), float(m[i][j][2])] 
      avg = float(mean(lst)) 
      newImage[i][j][0] = avg 
      newImage[i][j][1] = avg 
      newImage[i][j][2] = avg 
      newImage[i][j][3] = 1
  

plt.imshow(newImage)
```

```python
""" FLIP IMAGE """

img = np.array(Image.open('graph.png'))
print(type(img))


print(img.shape)


plt.imshow(np.flipud(img))
```

```python
""" CROP IMAGE """
  
w, h = m.shape[:2] 
xNew = int(w * 1 / 2) 
yNew = int(h * 1 / 2) 
newImage = np.zeros([xNew, yNew, 4]) 
  
print(w) 
print(h) 
  
for i in range(1, xNew): 
    for j in range(1, yNew): 
        newImage[i, j]= m[100 + i, 100 + j] 


plt.imshow(newImage)
```
